```python
full_data = pickle.load(open("loaded_data.pickle", "rb"))
```

```python
exam_data = full_data["exam_18"]
```

```python
std_samp = exam_data["std_1"]
```

```python
training_data = pd.DataFrame(pd.io.json.json_normalize(exam_data["std_1"]))
```

```python
training_data
```

```python
index = 2
while True:
    try:
        name = "std_" + str(index)
        training_data = training_data.append(pd.io.json.json_normalize(exam_data[name]))
        index += 1
    except KeyError:
        break
training_data
```

```python
training_data.columns[training_data.isnull().any()].tolist()
```

```python
training_data_na = training_data.fillna(0)
training_data_na
```

```python
from sklearn.cluster import KMeans
```

```python
cluster_model = KMeans(n_clusters = 2, random_state=0, n_jobs=-1)
```

```python
cluster_model.fit(training_data_na)
```

```python
predictions = []
```

```python

e_index = 20
while True:
    try:
        exam_data = full_data["exam_"+str(e_index)]
        test_data = pd.DataFrame(pd.io.json.json_normalize(exam_data["std_1"]))
        index = 2
        while True:
            try:
                name = "std_" + str(index)
                test_data = test_data.append(pd.io.json.json_normalize(exam_data[name]))
            except KeyError:
                p = cluster_model.predict(test_data.fillna(0))
                predictions.append(p)
                break
            index += 1
    except KeyError:
        break
    except:
        pass
    print(e_index)
    e_index += 1
```

```python
len(predictions)
```

```python
def check (predicts):
    accuracies = []
    for predicted in predicts:
        trues = 0
        for i in range(len(predicted)):
            if i < 94:
                if predicted[i] == 0:
                    trues += 1
            else:
                if predicted[i] == 1:
                    trues += 1
        accuracies.append(float(trues)/99)
    return accuracies
```

```python
check(predictions)
```
